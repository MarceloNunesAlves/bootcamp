```python
import numpy as np 
import pandas as pd
from pandas.plotting import scatter_matrix
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import transforms
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

sns.set_style('whitegrid', {"axes.grid" : False})
sns.set_context('notebook')
np.random.seed(42)

pd.set_option('display.max_columns', None)
```

```python
prod_renda_fixa = pd.read_csv("../repository/produtos_groups_renda_fixa.csv",sep=',')
prod_fundos = pd.read_csv("../repository/produtos_groups_fundos.csv",sep=',')
clientes = pd.read_csv("../repository/clientes_groups.csv",sep=',')

clientes['label_cli'] = clientes['label']
clientes = pd.DataFrame(clientes[['label_cli','ProdutoId']].groupby(['label_cli','ProdutoId']).count()).reset_index()
clientes
```

```python
joindata_renda_fixa = pd.merge(clientes, prod_renda_fixa, on='ProdutoId')
joindata_renda_fixa
```

```python
joindata_renda_fixa.to_csv('../repository/processado_renda_fixa.csv')
```

```python
joindata_fundos = pd.merge(clientes, prod_fundos, on='ProdutoId')
joindata_fundos
```

```python
joindata_fundos.to_csv('../repository/processado_fundos.csv')
```
